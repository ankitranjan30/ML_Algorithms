```python
#Store the Pearson Correlation in a dictionary, where the key is the user Id and the value is the coefficient
pearsonCorrelationDict = {}

#For every user group in our subset
for name, group in common_titles_by_group:
    #Let's start by sorting the input and current user group so the values aren't mixed up later on
    group = group.sort_values(by='Title')
    input_titles_for_recommender = input_titles_for_recommender.sort_values(by='Title')
    #Get the N for the formula
    nRatings = len(group)
    #Get the review scores for the movies that they both have in common
    temp_df = input_titles_for_recommender[input_titles_for_recommender['Title'].isin(group['Title'].tolist())]
    #And then store them in a temporary buffer variable in a list format to facilitate future calculations
    tempRatingList = temp_df['Your Rating'].tolist()
    #Let's also put the current user group reviews in a list format
    tempGroupList = group['Your Rating'].tolist()
    #Now let's calculate the pearson correlation between two users, so called, x and y
    Sxx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList),2)/float(nRatings)
    Syy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList)*sum(tempGroupList)/float(nRatings)
    
    #If the denominator is different than zero, then divide, else, 0 correlation.
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[name] = Sxy/sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[name] = 0
```

```python
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
pearsonDF.columns = ['similarityIndex']
pearsonDF['userId'] = pearsonDF.index
pearsonDF.index = range(len(pearsonDF))
pearsonDF.head()
```

```python
useful_features = ['Genres', 'Year', 'Directors', 'Runtime (mins)', 'Your Rating','User_ID']
feature_set_User_ID_1_genre_year_director_length = titles_of_User_ID_1[useful_features]
feature_set_User_ID_1_genre_year_director_length

feature_set_rem_users_genre_year_director_length = titles_of_users_apart_from_User_ID_1[useful_features]
feature_set_rem_users_genre_year_director_length

feature_set_all_users_genre_year_director_length = all_user_ratings[useful_features]
feature_set_all_users_genre_year_director_length
```

```python
a1 = feature_set_all_users_genre_year_director_length['Genres'].str.get_dummies(sep=",")[:no_of_titles_userID_1] # The first no_of_titles_userID_1 belongs to User1
a2 = feature_set_all_users_genre_year_director_length['Directors'].str.get_dummies(sep=",")[:no_of_titles_userID_1] # The first no_of_titles_userID_1 belongs to User1
#a3 = titles_of_User_ID_1[['Year', 'Release Date Integer']]

no_titles_rem_users = all_user_ratings.shape[0] - no_of_titles_userID_1

b1 = feature_set_all_users_genre_year_director_length['Genres'].str.get_dummies(sep=",")[no_of_titles_userID_1:].reset_index().drop(['index'], axis = 1) # The remaining belong to other Users
b2 = feature_set_all_users_genre_year_director_length['Directors'].str.get_dummies(sep=",")[no_of_titles_userID_1:].reset_index().drop(['index'], axis = 1) # The remaining belong to other Users
#b3 = titles_of_users_apart_from_User_ID_1[['Year', 'Release Date Integer']].reset_index().drop(['index'], axis = 1)



ratings_userID_1 = titles_of_User_ID_1['Your Rating']
ratings_rem_users = titles_of_users_apart_from_User_ID_1['Your Rating']

#Combining features
d1 = pd.concat([a1, a2], axis = 1) #UserID_1
d1

d2 = pd.concat([b1, b2], axis = 1) #rem_users
d2


weighted_user1 = np.dot(ratings_userID_1, d1) 
weighted_rem_users = np.dot(ratings_rem_users, d2)

#Normalizing weights
weighted_matrix_user1 = (weighted_user1/np.sum(weighted_user1)).reshape(len(weighted_user1), 1) 
weighted_matrix_rem_users = (weighted_rem_users/np.sum(weighted_rem_users)).reshape(len(weighted_rem_users), 1) 

print(a1.shape)
print(b1.shape)

```

```python
weighted_matrix_user1 = weighted_matrix_user1.T
```

```python
first_N = len(input_titles_for_recommender) #10
input_titles_for_recommender_1 = d2[:first_N]

# recommendation_matrix for titles not watched by UserID_1
recommendation_matrix = pd.DataFrame(np.array(np.sum((weighted_matrix_user1)*(input_titles_for_recommender_1), axis = 1)).reshape(len(input_titles_for_recommender), 1))
recommendation_matrix

input_titles_for_recommender = input_titles_for_recommender.reset_index().drop(['index'], axis = 1)
input_titles_for_recommender
recommendation_matrix = pd.concat([recommendation_matrix, input_titles_for_recommender], axis = 1).rename(columns={0: "Weighted_Average"})

recommendation_matrix
```

```python
# Sorting by most recommended to least
recommendation_matrix = recommendation_matrix.sort_values(by=['Weighted_Average'], ascending = False)
```
